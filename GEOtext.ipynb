```python
df.set_index("ID_REF",inplace = True)
print(df)
```

```python
import matplotlib.pyplot as plt
df.boxplot(figsize = (10,6))
plt.xticks(rotation = 45)
plt.title("Boxplot of expression values for 6 samples")
plt.show()
```

因为是同一种细胞系的重复，所以没有明显差异

```python
#画热图
import seaborn as sns
#提前前50个元素
df_top50 = df.head(50)
plt.figure(figsize = (6,8))
sns.heatmap(df_top50,cmap = "coolwarm",
           center = 0 ,xticklabels = True,
           yticklabels = True)
plt.title("Heatmap of top 50 genes (rows) across 6 samples")
plt.show()
```

```python
#找出变化最大的基因
genevariance = df.var(axis = 1)
genevariance.head()
```

```python
top_10genes = genevariance.sort_values(ascending = False).head(10)
print(top_10genes)
```

```python
top_expr = df.loc[top_10genes.index]
print(top_expr)
```

```python
import matplotlib.pyplot as plt
plt.rcParams["font.sans-serif"] = "SimHei"
plt.rcParams["axes.unicode_minus"] = False
plt.figure(figsize = [10,6])
#提前样本，对每个基因画一条线
for gene_id in top_expr.index:
    plt.plot(top_expr.columns,top_expr.loc[gene_id],marker = "o",label = str(gene_id))
plt.xlabel("样本")
plt.ylabel("表达值（log2）")
plt.title("表达变化最大的前十个基因在不同样本中的表达趋势")
plt.xticks(rotation = 45)
plt.legend(bbox_to_anchor =(1.05,1),loc = "upper left")
plt.tight_layout()
plt.show()
```
